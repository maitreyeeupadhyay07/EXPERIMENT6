```python
##EXPERIMENT 6 CONTINUED
#MAITREYEE UPADHYAY 25070123071

# WRITE A PYTHON PROGRAM TO CALCULATE THE GROSS SALARY OF AN EMPLOYEE BASED
#IF BASIC<= 10000: HRA = 20%, DA = 80%
#IF BASIC <= 20000: HRA = 25%, DA = 90%
#OTHERWISE: HRA = 30%, DA = 95%

basic = float(input("Enter basic salary: "))
if basic <= 10000:
    hra = basic * 0.2
    da = basic * 0.90
elif basic <= 20000:
    hra = basic*0.25
    da = basic*0.90
else:
    hra = basic*0.3
    da = basic*0.95

gross = basic + hra + da
print("Gross salary: ", gross)
```

```python
#7. WRITE A PYTHON PROGRAM TO CALCULATE INCOME TAX BASED ON ANNUAL INCOME
# UP TO RS 2,50,000: NO TAX
# RS 2,50,001 - RS 5,00,000: 5%
# RS 5,00,001 - RS 10,00,000: 20%
#ABOVE RS 10,00,000: 30%

income = float(input("Enter annual income: "))
if income <= 250000:
   income_tax = 0
elif income <= 500000:
   income_tax = (income - 250000) * 0.05
elif income <= 1000000:
   income_tax = (250000*0.05) + (income - 500000) * 0.20
else:
   income_tax = (250000*0.05) + (500000*0.20) + (income - 1000000) * 0.30

   print("Income tax: ", income_tax)
```
